```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
salf=pd.read_csv("/content/plastic-waste-generation.csv")
```

```python
salf.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_Entity = LabelEncoder()

salf['Entity'] = le_Entity.fit_transform(salf['Entity'])
```

```python
salf.head()
```

```python
inp=salf[['Entity','Year']]
out=salf["Plastic waste generation (tonnes, total)"]
```

```python
LR=LogisticRegression()
```

```python
LR.fit(inp, out)
```

```python
salf['Entity'] = le_Entity.inverse_transform(salf['Entity'])
```

```python
Entity = int(input("Enter the country's encoded ID : "))
year_input = int(input("Enter the Year: "))
pred=LR.predict([[Entity,year_input]])
print(pred)
```
