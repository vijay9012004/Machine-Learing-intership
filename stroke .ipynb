```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/panda/healthcare-dataset-stroke-data (1).csv")

# Select useful numeric columns
X = df[["stroke"]]
y = df["avg_glucose_level"]

# Train the model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Plot
plt.figure(figsize=(6,4))
plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_pred, label="Regression Line")
plt.xlabel("stroke")
plt.ylabel("Average Glucose Level")
plt.title("Linear Regression: stroke vs Glucose Level")
plt.legend()
plt.grid(True)
plt.show()
```
